```python
import pandas as pd
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('data/datossub.csv')
data=data.set_index('index')

data = data.sample(100, random_state=987)
print(data.tail())

G = nx.from_pandas_edgelist(data, 'Follower', 'Target', create_using=nx.DiGraph())
nx.draw(G, with_labels=True, node_size=50, alpha=0.5, arrows=True)
```

```python
B = nx.Graph()
B.add_nodes_from(data['Follower'],bipartite=0)
B.add_nodes_from(data['Target'],bipartite=1)
B.add_edges_from(data)
# Graficar la red bipartita
nx.draw_networkx(B)
plt.axis("off")
plt.show()
```

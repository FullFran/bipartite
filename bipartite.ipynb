```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
B = nx.Graph()
B.add_nodes_from([1, 2, 3], bipartite=0)  # los nodos 1, 2, y 3 pertenecen a la partición 0
B.add_nodes_from(['a', 'b', 'c'], bipartite=1)  # los nodos 'a', 'b', y 'c' pertenecen a la partición 1


B.add_edges_from([(1, 'a'), (1, 'b'), (2, 'b'), (3, 'c')])
```

```python
pos = nx.bipartite_layout(B, [1, 2, 3]) # Obtener las coordenadas de los nodos

nx.draw(B, pos, with_labels=True) # Dibujar los nodos y las conexiones
plt.show() # Mostrar la visualización
```

```python
# Crear la red bipartita
B = nx.Graph()
users = ["user1", "user2", "user3"]
words = ["python", "data", "networkx", "tweepy", "twitter"]
for user in users:
    B.add_node(user, bipartite=0)
for word in words:
    B.add_node(word, bipartite=1)
B.add_edge("user1", "python")
B.add_edge("user1", "data")
B.add_edge("user2", "data")
B.add_edge("user2", "networkx")
B.add_edge("user3", "tweepy")
B.add_edge("user3", "twitter")

# Definir los colores de los nodos y las posiciones para graficar
color_map = []
pos = {}
for i, user in enumerate(users):
    color_map.append("blue")
    pos[user] = (0, i)
for i, word in enumerate(words):
    color_map.append("green")
    pos[word] = (1, i)

# Graficar la red bipartita
nx.draw_networkx(B, pos=pos, node_color=color_map)
plt.axis("off")
plt.show()
```

```python
# Datos ficticios
data = {
    "user1": {"python": 3, "data": 1, "pandas": 2},
    "user2": {"data": 2, "networkx": 1, "numpy": 2},
    "user3": {"tweepy": 1, "twitter": 2, "matplotlib": 3}
}

# Crear la red bipartita
B = nx.Graph()
users = ["user1", "user2", "user3"]
words = set()
for user in users:
    B.add_node(user, bipartite=0)
    for word, weight in data[user].items():
        B.add_node(word, bipartite=1)
        words.add(word)
        B.add_edge(user, word, weight=weight)

# Definir los colores de los nodos y las posiciones para graficar
color_map = []
pos = {}
for i, user in enumerate(users):
    color_map.append("blue")
    pos[user] = (0, i)
for i, word in enumerate(words):
    color_map.append("green")
    pos[word] = (1, i)

# Definir los pesos de las aristas para graficar
weights = [B[u][v]["weight"] for u, v in B.edges()]

# Graficar la red bipartita
nx.draw_networkx(B, pos=pos, node_color=color_map, width=weights, edge_color="gray", font_size=10, font_weight="bold", with_labels=True)
plt.axis("off")
plt.show()
```

```python
from networkx.algorithms import bipartite

# Obtener la proyección de la red en el conjunto de palabras
words_projection = bipartite.projected_graph(B, words)

# Obtener la proyección de la red en el conjunto de usuarios
users_projection = bipartite.projected_graph(B, users)

# Imprimir el número de nodos y aristas en cada proyección
print("Proyección de palabras:")
print("Número de nodos:", len(words_projection.nodes()))
print("Número de aristas:", len(words_projection.edges()))
print("\nProyección de usuarios:")
print("Número de nodos:", len(users_projection.nodes()))
print("Número de aristas:", len(users_projection.edges()))
```

```python
# Obtener la proyección de la red en el conjunto de palabras
words_projection = bipartite.projected_graph(B, words)

# Obtener la proyección de la red en el conjunto de usuarios
users_projection = bipartite.projected_graph(B, users)

# Graficar la proyección de palabras
pos = nx.spring_layout(words_projection, seed=42)
nx.draw(words_projection, pos, with_labels=True)
nx.draw_networkx_edge_labels(words_projection, pos, edge_labels=nx.get_edge_attributes(words_projection, 'weight'))
plt.title('Proyección de palabras')
plt.show()

# Graficar la proyección de usuarios
pos = nx.spring_layout(users_projection, seed=42)
nx.draw(users_projection, pos, with_labels=True)
nx.draw_networkx_edge_labels(users_projection, pos, edge_labels=nx.get_edge_attributes(users_projection, 'weight'))
plt.title('Proyección de usuarios')
plt.show()
```

```python
# Creamos la red bipartita
B = nx.Graph()

# Creamos la lista de usuarios y la lista de palabras
users = ['user1', 'user2', 'user3', 'user4', 'user5', 'user6']
words = ['word1', 'word2', 'word3', 'word4', 'word5', 'word6', 'word7', 'word8']

# Añadimos los nodos de usuarios y palabras a la red bipartita
B.add_nodes_from(users, bipartite=0)
B.add_nodes_from(words, bipartite=1)

# Añadimos los pesos a las aristas de la red
edges = [('user1', 'word1', {'weight': 3}), ('user1', 'word2', {'weight': 1}), ('user1', 'word3', {'weight': 2}),
         ('user2', 'word2', {'weight': 1}), ('user2', 'word4', {'weight': 2}), ('user2', 'word5', {'weight': 1}),
         ('user3', 'word1', {'weight': 2}), ('user3', 'word3', {'weight': 3}), ('user3', 'word6', {'weight': 1}),
         ('user4', 'word2', {'weight': 2}), ('user4', 'word5', {'weight': 2}), ('user4', 'word7', {'weight': 1}),
         ('user5', 'word4', {'weight': 1}), ('user5', 'word7', {'weight': 2}), ('user5', 'word8', {'weight': 3}),
         ('user6', 'word3', {'weight': 1}), ('user6', 'word6', {'weight': 3}), ('user6', 'word8', {'weight': 2})]

# Añadimos las aristas a la red bipartita
B.add_edges_from(edges)

# Obtenemos la proyección de usuarios
user_projection = nx.bipartite.weighted_projected_graph(B, users)

# Obtenemos la proyección de palabras
word_projection = nx.bipartite.weighted_projected_graph(B, words)

# Dibujamos la red bipartita
pos = {n: (0, i) for i, n in enumerate(users)}
pos.update({n: (1, i) for i, n in enumerate(words)})
nx.draw_networkx(B, pos=pos, with_labels=True, node_color='lightblue', edge_color='gray')

# Dibujamos la proyección de usuarios
plt.figure()
nx.draw_networkx(user_projection, with_labels=True, node_color='lightblue', edge_color='gray')

# Dibujamos la proyección de palabras
plt.figure()
nx.draw_networkx(word_projection, with_labels=True, node_color='lightgreen', edge_color='gray')

# Mostramos los gráficos
plt.show()
```

```python
import random

# Generar datos de usuarios y palabras aleatorios
users = [f'User{i+1}' for i in range(10)]
words = [f'Word{j+1}' for j in range(5)]
user_word_counts = {user: {word: random.randint(0, 100) for word in words} for user in users}

# Crear grafo bipartito
G = nx.Graph()
G.add_nodes_from(users, bipartite=0)
G.add_nodes_from(words, bipartite=1)

# Agregar bordes con pesos
for user, word_counts in user_word_counts.items():
    for word, count in word_counts.items():
        G.add_edge(user, word, weight=count)

# Graficar red bipartita
pos = {node: (0, i) for i, node in enumerate(users)}
pos.update({node: (1, i) for i, node in enumerate(words)})
nx.draw(G, pos, with_labels=True, font_size=10, node_color='lightblue', node_size=500, width=2,
        edge_color=[G[u][v]['weight'] for u, v in G.edges()], edge_cmap=plt.cm.Blues)
plt.show()

# Graficar red bipartita
pos = {node: (0, i) for i, node in enumerate(users)}
pos.update({node: (1, i) for i, node in enumerate(words)})
nx.draw(G, pos, with_labels=True, font_size=10, node_color='lightblue', node_size=500,
        width=[d['weight']/100 for (u, v, d) in G.edges(data=True)])
plt.show()

print(user_word_counts)
```

```python
user_proj = nx.bipartite.projected_graph(G, users)

user_edge_weights = []
for u, v in user_proj.edges():
    weight = sum([G[u][w]['weight']/100 for w in G[u] if w != v])
    user_edge_weights.append(weight)

nx.draw(user_proj, with_labels=True, font_size=10, node_color='lightblue', node_size=500,
        width=user_edge_weights, edge_color=user_edge_weights, edge_cmap=plt.cm.Blues)

plt.show()


word_proj = nx.bipartite.projected_graph(G, words)
word_edge_weights = []
for u, v in word_proj.edges():
    weight = sum([G[u][w]['weight']/100 for w in G[u] if w != v])
    word_edge_weights.append(weight)
nx.draw(word_proj, with_labels=True, font_size=10, node_color='lightblue', node_size=500,
        width=word_edge_weights, edge_color=word_edge_weights, edge_cmap=plt.cm.Blues)
plt.show()
print(word_edge_weights)


nx.write_gexf(user_proj, "proyeccion_users.gexf")

nx.write_gexf(word_proj, "proyeccion_palabras.gexf")
```

```python

# Obtener la lista de grados de los nodos de la partición de los usuarios
user_degrees = [d for n, d in G.degree() if G.nodes[n]['bipartite'] == 0]

# Graficar la distribución de grados
plt.hist(user_degrees, bins=range(12), align='left', rwidth=0.8, color='lightblue')
plt.xticks(range(11))
plt.xlabel('Grado')
plt.ylabel('Número de nodos')
plt.title('Distribución de grados de la partición de usuarios')
plt.show()
```

```python
from collections import Counter

# Calcular el grado ponderado de cada nodo
weighted_degree = dict(G.degree(weight='weight'))

# Contar cuántos nodos tienen el mismo grado
degree_counts = Counter(weighted_degree.values())

# Graficar la distribución de grados
plt.bar(degree_counts.keys(), degree_counts.values())
plt.xlabel('Grado ponderado')
plt.ylabel('Número de nodos')
plt.show()
```
